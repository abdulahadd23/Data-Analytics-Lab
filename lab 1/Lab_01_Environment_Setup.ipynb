## 1.4 — Basic Python Operations

````python
# Arithmetic Operations
a, b = 25, 7
print("=== Arithmetic Operations ===")
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b:.4f}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
````

````python
# String Operations
name = "Data Analytics"
print("=== String Operations ===")
print(f"Original: {name}")
print(f"Upper: {name.upper()}")
print(f"Lower: {name.lower()}")
print(f"Length: {len(name)}")
print(f"Split: {name.split()}")
print(f"Replace: {name.replace('Analytics', 'Science')}")
print(f"Starts with 'Data': {name.startswith('Data')}")
````

````python
# List Operations
numbers = [15, 8, 23, 4, 42, 16, 7]
print("=== List Operations ===")
print(f"Original: {numbers}")
print(f"Sorted: {sorted(numbers)}")
print(f"Reversed: {numbers[::-1]}")
print(f"Sum: {sum(numbers)}")
print(f"Min: {min(numbers)}, Max: {max(numbers)}")
print(f"Sliced [2:5]: {numbers[2:5]}")
````

````python
# Dictionary Operations
student = {
    "name": "Abdul Ahad",
    "course": "Data Analytics",
    "semester": 4,
    "gpa": 3.5,
    "skills": ["Python", "SQL", "Visualization"]
}

print("=== Dictionary Operations ===")
for key, value in student.items():
    print(f"  {key}: {value}")

print(f"\nKeys: {list(student.keys())}")
print(f"Has 'gpa': {'gpa' in student}")
````

## 1.6 — First Data Visualization

````python
# Quick visualization to verify matplotlib + seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")

categories = ['Healthcare', 'Finance', 'Retail', 'Manufacturing', 'Education']
adoption = [78, 92, 85, 67, 55]
colors = sns.color_palette("viridis", len(categories))

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(categories, adoption, color=colors, edgecolor='white', height=0.6)
ax.set_xlabel('AI/Analytics Adoption Rate (%)', fontsize=12)
ax.set_title('Data Analytics Adoption Across Industries', fontsize=14, fontweight='bold')
ax.set_xlim(0, 100)

for bar, val in zip(bars, adoption):
    ax.text(bar.get_width() + 1.5, bar.get_y() + bar.get_height()/2,
            f'{val}%', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig('industry_adoption.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nVisualization saved successfully!")
````
