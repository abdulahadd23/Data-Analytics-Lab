## 2.1 — Python Data Structures

### Lists

```python
# Lists — ordered, mutable collection
grades = [85, 92, 78, 95, 88, 76, 91, 83, 97, 72]
print("=== List Operations ===")
print(f"Grades: {grades}")
print(f"Length: {len(grades)}")
print(f"Average: {sum(grades)/len(grades):.2f}")
print(f"Highest: {max(grades)}")
print(f"Lowest: {min(grades)}")

# Append and extend
grades.append(89)
print(f"After append(89): {grades}")

# Sorting
print(f"Sorted (asc): {sorted(grades)}")
print(f"Sorted (desc): {sorted(grades, reverse=True)}")
print(f"Top 3: {sorted(grades, reverse=True)[:3]}")
```

### Dictionaries

```python
# Dictionaries — key-value pairs
student_records = {
    "STU001": {"name": "Ali Khan", "grade": 85, "course": "Data Analytics"},
    "STU002": {"name": "Sara Ahmed", "grade": 92, "course": "Data Analytics"},
    "STU003": {"name": "Usman Tariq", "grade": 78, "course": "Machine Learning"},
    "STU004": {"name": "Fatima Noor", "grade": 95, "course": "Data Analytics"},
    "STU005": {"name": "Hassan Raza", "grade": 88, "course": "Machine Learning"},
}

print("=== Student Records ===")
for sid, info in student_records.items():
    print(f"  {sid}: {info['name']} — Grade: {info['grade']}, Course: {info['course']}")

# Filter by course
da_students = {k: v for k, v in student_records.items() if v['course'] == 'Data Analytics'}
print(f"\nData Analytics students: {len(da_students)}")

# Average grade
avg = sum(v['grade'] for v in student_records.values()) / len(student_records)
print(f"Average grade: {avg:.2f}")
```

### Tuples & Sets

```python
# Tuples — immutable sequences
coordinates = (33.6844, 73.0479)  # Islamabad
print(f"Coordinates: {coordinates}")
print(f"Latitude: {coordinates[0]}, Longitude: {coordinates[1]}")

# Named tuple pattern
from collections import namedtuple
City = namedtuple('City', ['name', 'lat', 'lon', 'population'])
cities = [
    City('Karachi', 24.8607, 67.0011, 14910352),
    City('Lahore', 31.5204, 74.3587, 11126285),
    City('Islamabad', 33.6844, 73.0479, 1014825),
]
print("\n=== Pakistani Cities ===")
for c in cities:
    print(f"  {c.name}: ({c.lat}, {c.lon}) — Pop: {c.population:,}")

# Sets — unique elements
skills_team1 = {'Python', 'SQL', 'Excel', 'Tableau'}
skills_team2 = {'Python', 'R', 'Power BI', 'SQL'}
print(f"\n=== Set Operations ===")
print(f"Team 1: {skills_team1}")
print(f"Team 2: {skills_team2}")
print(f"Union: {skills_team1 | skills_team2}")
print(f"Intersection: {skills_team1 & skills_team2}")
print(f"Team 1 only: {skills_team1 - skills_team2}")
```

## 2.2 — Functions for Data Processing

```python
def calculate_statistics(data, label="Dataset"):
    """Calculate comprehensive statistics for a numeric list."""
    import statistics
    n = len(data)
    stats = {
        'count': n,
        'mean': sum(data) / n,
        'median': statistics.median(data),
        'std_dev': statistics.stdev(data),
        'min': min(data),
        'max': max(data),
        'range': max(data) - min(data),
        'q1': statistics.quantiles(data, n=4)[0],
        'q3': statistics.quantiles(data, n=4)[2],
    }
    stats['iqr'] = stats['q3'] - stats['q1']

    print(f"=== Statistics for {label} ===")
    for key, val in stats.items():
        print(f"  {key:>10}: {val:.2f}" if isinstance(val, float) else f"  {key:>10}: {val}")
    return stats

# Test with sample data
test_scores = [72, 85, 91, 68, 77, 94, 83, 79, 88, 65, 96, 73, 81, 90, 76]
stats = calculate_statistics(test_scores, "Student Test Scores")
```

```python
def grade_classifier(score):
    """Classify a numeric score into a letter grade."""
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'

def process_student_grades(students):
    """Process a list of student dicts and add computed fields."""
    for s in students:
        s['letter_grade'] = grade_classifier(s['score'])
        s['pass_fail'] = 'Pass' if s['score'] >= 60 else 'Fail'
    return students

students = [
    {"name": "Ahmed", "score": 92},
    {"name": "Bilal", "score": 78},
    {"name": "Chand", "score": 55},
    {"name": "Danish", "score": 88},
    {"name": "Esha", "score": 95},
    {"name": "Farhan", "score": 63},
]

processed = process_student_grades(students)
print(f"{'Name':<10} {'Score':>5} {'Grade':>5} {'Status':>6}")
print("-" * 30)
for s in processed:
    print(f"{s['name']:<10} {s['score']:>5} {s['letter_grade']:>5} {s['pass_fail']:>6}")
```

## 2.4 — List Comprehensions

```python
# List comprehension examples for data transformation
numbers = list(range(1, 21))

# Basic transformations
squares = [x**2 for x in numbers]
evens = [x for x in numbers if x % 2 == 0]
print(f"Numbers: {numbers}")
print(f"Squares: {squares}")
print(f"Evens: {evens}")

# Practical: Grade classification
scores = [92, 78, 55, 88, 95, 63, 71, 84, 49, 77]
grades = [(s, grade_classifier(s)) for s in scores]
print(f"\nGrades: {grades}")

# Nested comprehension — flatten a matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print(f"\nMatrix flattened: {flat}")

# Dictionary comprehension
score_map = {f"Student_{i+1}": s for i, s in enumerate(scores)}
passing = {k: v for k, v in score_map.items() if v >= 60}
print(f"\nPassing students: {passing}")

# Performance: list comp vs loop
import time

n = 1_000_000
start = time.time()
result_loop = []
for i in range(n):
    result_loop.append(i ** 2)
loop_time = time.time() - start

start = time.time()
result_comp = [i ** 2 for i in range(n)]
comp_time = time.time() - start

print(f"\n=== Performance Comparison (n={n:,}) ===")
print(f"  For loop:          {loop_time:.4f}s")
print(f"  List comprehension: {comp_time:.4f}s")
print(f"  Speedup:           {loop_time/comp_time:.2f}x")
```

## 2.5 — Error Handling in Data Processing

```python
def safe_read_csv(filepath):
    """Safely read a CSV file with comprehensive error handling."""
    try:
        with open(filepath, 'r') as f:
            reader = csv.DictReader(f)
            data = list(reader)
        print(f"Successfully read {len(data)} records from '{filepath}'")
        return data
    except FileNotFoundError:
        print(f"Error: File '{filepath}' not found.")
        return None
    except PermissionError:
        print(f"Error: No permission to read '{filepath}'.")
        return None
    except csv.Error as e:
        print(f"CSV Error: {e}")
        return None

def safe_convert(value, target_type=float, default=0):
    """Safely convert a value to a target type."""
    try:
        return target_type(value)
    except (ValueError, TypeError):
        return default

# Demonstration
print("=== Error Handling Demo ===\n")

# 1. Valid file
data = safe_read_csv('sales_data.csv')

# 2. Missing file
data = safe_read_csv('nonexistent.csv')

# 3. Safe type conversion
test_values = ["42", "3.14", "abc", None, "", "100"]
print(f"\nSafe conversions:")
for v in test_values:
    result = safe_convert(v)
    print(f"  '{v}' → {result}")

# 4. Custom exception
class DataValidationError(Exception):
    pass

def validate_score(score):
    if not isinstance(score, (int, float)):
        raise DataValidationError(f"Score must be numeric, got {type(score).__name__}")
    if not 0 <= score <= 100:
        raise DataValidationError(f"Score {score} out of range [0, 100]")
    return True

test_scores_mixed = [85, 110, "abc", -5, 92, None]
print("\nScore validation:")
for s in test_scores_mixed:
    try:
        validate_score(s)
        print(f"  {s!r:>6} → Valid")
    except DataValidationError as e:
        print(f"  {s!r:>6} → Invalid: {e}")
    except TypeError:
        print(f"  {s!r:>6} → Invalid: unsupported type")
```
